```python
import pandas as pd

train = pd.read_csv('./Parking_data/train.csv')
test = pd.read_csv('./Parking_data/test.csv')
sub = pd.read_csv('./Parking_data/sample_submission.csv')
age = pd.read_csv('./Parking_data/age_gender_info.csv')

print(train.shape, test.shape, sub.shape, age.shape)
train
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
sel = ['전용면적']
X = train[sel]
y = train['등록차량수']
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=24)

Line = LinearRegression().fit(X_train, y_train)
pred = Line.predict(X_test)

df = {'실제값':y_test, "예측값":pred, '오차': y_test-pred}
df = pd.DataFrame(df)
df
```
